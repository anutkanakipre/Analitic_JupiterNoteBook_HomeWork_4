```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head(3000)
```

### 1.2 Изучите стоимости недвижимости


```python
plt.figure(figsize=(6, 4))
plt.hist(df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

### 1.3 Изучите распределение квадратуры жилой


```python
plt.figure(figsize=(6, 7))
sns.histplot(df['sqft_living'], bins=20)
plt.title('Распределение квадратуры жилой')
plt.xlabel('Кв. метры жилые')
plt.ylabel('Количество');
```

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(6, 7))
sns.histplot(df['yr_built'], bins=20)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

## Задача 2
### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы


```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(5, 3))
plt.pie(data, autopct='%.1f%%')
plt.legend(['no', 'yes']);
```

### 2.2 Изучите распределение этажей домов


```python
data = df['floors'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%', labels=data.index)
plt.legend(['Количесиво этажей']);
```

### 2.2 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()

plt.figure(figsize=(6, 5))
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

## 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head(3000)
```

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg'); # Чем выше качество дома, тем выше его стоимость.
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg'); # Чем выше жилая площадь, тем выше стоимость.
```

```python
sns.jointplot(x=df['price'], y=df['waterfront'], kind='reg'); # Цена растет при при наличии вида на набережную
```

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Зависимость стоимости дома от оценки качества');

#  Наибольшая стоимость у домов с оценкой 13.
```

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Зависимосит стоимости домов от количества этажей');

#  Наибольшая стоимость у домов, в которых 2,5 этажа.
```
